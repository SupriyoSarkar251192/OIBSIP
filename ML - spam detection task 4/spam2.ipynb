## Importing necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
nltk.download('stopwords')
```

```python
nltk.download('averaged_perceptron_tagger')
```

## Loading the dataset

```python
df = pd.read_csv("smsspam.txt", sep='\t',names=["label","message"])
df1 = df.copy()
df1.head()
```

## Cleaning Messages

```python
stemmer = PorterStemmer()
```

```python
cleaned_messages = []
for i in df1['message']:
    msg1=re.sub('[^a-zA-z]',' ',i)
    msg1 = msg1.lower()
    msg1 = word_tokenize(msg1)
    msg1 = [stemmer.stem(j) for j in msg1 if j not in stopwords.words('english')]
    msg1 = ' '.join(msg1)
    cleaned_messages.append(msg1)
cleaned_messages
```

```python
df1['cleaned_messages'] = cleaned_messages
df1.head()
```

## Encoding

```python
encoder = LabelEncoder()
tfidf = TfidfVectorizer(max_features=3000)

y = encoder.fit_transform(df1['label'])
x = tfidf.fit_transform(df1['cleaned_messages'])
```

## Splitting Data for Model Training

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
clf = RandomForestClassifier()
params = {
    'max_depth': [1, 2, 3, 5, 10, None],
    'n_estimators': [50, 100, 200, 300],
    'criterion': ['gini', 'entropy']
}

model = GridSearchCV(clf, param_grid=params, cv = 5, verbose=3, scoring='accuracy')
model.fit(x_train, y_train)
```

```python
model.best_score_, model.best_params_
```

## Prediction and Accuracy

```python
y_pred = model.predict(x_test)
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Building Spam Detector

```python
def spam_detector(msg, dataset = df):
    
    import nltk
    import re
    from nltk.corpus import stopwords
    from nltk.tokenize import word_tokenize
    from nltk.stem import PorterStemmer

    from sklearn.preprocessing import LabelEncoder
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.ensemble import RandomForestClassifier
    
    nltk.download('punkt_tab')
    nltk.download('stopwords')
    nltk.download('averaged_perceptron_tagger')
    
    stemmer = PorterStemmer()
    
    inp_msg =re.sub('[^a-zA-z]',' ',msg)
    inp_msg = inp_msg.lower()
    inp_msg = word_tokenize(inp_msg)
    inp_msg = [stemmer.stem(j) for j in inp_msg if j not in stopwords.words('english')]
    inp_msg = ' '.join(msg)
    inp_msg = pd.DataFrame([inp_msg], columns=['msg'])
    
    cleaned_messages = []
    for i in dataset['message']:
        msg1=re.sub('[^a-zA-z]',' ',i)
        msg1 = msg1.lower()
        msg1 = word_tokenize(msg1)
        msg1 = [stemmer.stem(j) for j in msg1 if j not in stopwords.words('english')]
        msg1 = ' '.join(msg1)
        cleaned_messages.append(msg1)
        
    dataset['cleaned_messages'] = cleaned_messages
    
    encoder = LabelEncoder()
    tfidf = TfidfVectorizer(max_features=3000)

    y = encoder.fit_transform(dataset['label'])
    x = tfidf.fit_transform(dataset['cleaned_messages'])
    
    clf = RandomForestClassifier(criterion='gini', max_depth= None, n_estimators = 300, random_state=42)
    clf.fit(x, y)
    pred = clf.predict(tfidf.transform(inp_msg['msg']))
    
    if pred == 0:
        return 'ham'
    else:
        return 'spam'
```

```python
txt = "Keeping in mind the deteriorating weather condition, guardians are suggested to collect their wards early today according to their convenient time. All classes will remain suspended on 24.10.2024 and 25.10.2024 due to the upcoming cyclone."
spam_detector(txt)
```
